```python
thirty_rest = np.loadtxt("30rest.txt", float, delimiter = ",")
thirty_rest = np.reshape(thirty_rest, (30,66,1200))
```

```python
tf.reset_default_graph
```

```python
from random import randint 
def sample(all_scans):
     return all_scans[randint(0,29)]
```

```python
rest = tf.placeholder(tf.float32, shape = (66,1199))
```

```python
dr = tf.placeholder(tf.float32, shape = (66, 1199))
```

```python
dt = tf.constant(.72, dtype = tf.dtypes.float32)
```

```python
id = tf.constant(np.identity(66), dtype = tf.dtypes.float32)
```

```python
weight_matrix = np.random.randn(66,66)
weight_matrix = np.add(weight_matrix, weight_matrix.T)
weight_matrix = np.multiply(weight_matrix, .5)
diagonal = np.diag(np.diag(weight_matrix))
weight_matrix = np.subtract(weight_matrix, diagonal)
np.allclose(weight_matrix, weight_matrix.T, rtol = 1e-05, atol = 1e-08 ) #determines whether weight_matrix is symmetric
```

```python
weight_matrix = tf.Variable(weight_matrix, dtype = tf.dtypes.float32)
```

```python
weight_minus_id = tf.subtract(weight_matrix, id)
predicted = tf.linalg.matmul(weight_minus_id, rest)
predicted = tf.math.multiply(predicted, dt)
```

```python
loss = tf.losses.mean_squared_error(dr, predicted)
```

```python
optimizer = tf.train.AdamOptimizer(learning_rate = 1)
```

```python
init = tf.initialize_all_variables()
sess = tf.Session()
sess.run(init)
for i in range(10000):
    random_sample = sample(thirty_rest)
    current_frame = np.delete(random_sample, 1199, 1)
    next_step = np.delete(random_sample, 0, 1)
    change = np.subtract(current_frame, next_step)
    [my_weights, session_loss] = sess.run(fetches = [weight_matrix, loss], feed_dict = {rest: current_frame, dr: change})
    if i%100 == 0:
        print(session_loss, my_weights)
```
